```python
import matplotlib.pyplot as plt
import numba as nb
import numpy as np
import time
from sklearn.model_selection import KFold
from qosa import AveragedCheckFunction
from qosa import cross_validate
```

```python
@nb.njit("float64[:](float64[:,:], float64[:])", nogil=True, cache=False, parallel=True)
def _averaged_check_function_alpha_array(u, alpha):
    """
    Definition of the check function also called pinball loss function.
    """
    n_alphas = alpha.shape[0]
    n_samples = u.shape[0]

    check_function = np.empty((n_alphas, n_samples), dtype=np.float64)
    for i in nb.prange(n_samples):
        for j in range(n_alphas):
            check_function[j,i] = u[i,j]*(alpha[j] - (u[i,j] < 0.))

    averaged_check_function = np.empty((n_alphas), dtype=np.float64)
    for i in nb.prange(n_alphas):
        averaged_check_function[i] = check_function[i,:].mean()

    return averaged_check_function
```

### Test to find the good parameter with the inflexion point

```python
n_sample = 10**4
X = np.random.exponential(size = (n_sample, 2))
y = X[:,0] - X[:,1]
X = X[:,0].reshape(-1,1)

alpha = np.array([0.1, 0.5, 0.7])
min_samples_leaf = np.unique(np.logspace(np.log10(5), np.log10(1500), 20, endpoint=True, dtype=int))
used_bootstrap_samples = False
shuffle = True
n_estimators = 100
n_splits = 3
n_alphas = alpha.shape[0]

def check_function_error(estimator, X, y):
    alpha_quantile = np.percentile(y, q=alpha*100)
    denominator = _averaged_check_function_alpha_array(
                                                y.reshape(-1,1)-alpha_quantile,
                                                alpha)
    averaged_check_function = estimator.predict(alpha,
                                                used_bootstrap_samples=used_bootstrap_samples)
    return 1 - averaged_check_function[:,0]/denominator

n_leaf = min_samples_leaf.shape[0]
cross_val_values = np.zeros((n_leaf, n_splits, n_alphas), dtype=np.float64)
cross_val_mean_values = np.zeros((n_leaf, n_alphas), dtype=np.float64)

random_state = np.arange(min_samples_leaf.shape[0])
```

```python
%%time

for i in range(n_leaf):
    cv = KFold(n_splits=n_splits, shuffle=shuffle, random_state=random_state[3])
    cross_val_values[i,:,:] = cross_validate(
                                estimator=AveragedCheckFunction(
                                                    n_estimators=n_estimators, 
                                                    min_samples_leaf=min_samples_leaf[i],
                                                    random_state=random_state[3]), 
                                X=X,
                                y=y,
                                scoring=check_function_error,
                                cv=cv)['test_score']
    
    cross_val_mean_values[i,:] = cross_val_values[i,:,:].mean(axis=0)
```

```python
fig, axes = plt.subplots(figsize=(10,8))

axes.plot(min_samples_leaf, cross_val_mean_values, linewidth=2)
axes.legend(labels=(tuple([r'$ \alpha = %.2f $' % (i,) for i in alpha])))
```

```python
from scipy.optimize import curve_fit

def inflection_points(x_points, y_points):   
    # Take the second order finite difference of the curve with respect to the
    # defined coordinate system
    finite_difference_2 = np.gradient(np.gradient(y_points, x_points), x_points)
    
    # Compare the product of all neighboring pairs of points in the second derivative
    # If a pair of points has a negative product, then the second derivative changes sign
    # at one of those points, signalling an inflection point
    is_inflection_point = [finite_difference_2[i] * finite_difference_2[i + 1] <= 0 for i 
                           in range(len(finite_difference_2) - 1)]
    
    inflection_point_indices = np.array([i for i, b in enumerate(is_inflection_point) if b])
    return finite_difference_2, x_points[inflection_point_indices + 1][0]
```

```python
values_inflection_points = np.zeros(n_alphas)
for i in range(alpha.size):
    _, values_inflection_points[i] = inflection_points(min_samples_leaf, cross_val_mean_values[:,i])
```
